### Part I. Exploratory Data Analysis

Let's load some necessary modules.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

Now that we have the necessary modules, we can take a look at the data.

```python
#Loading the data
df = pd.read_json(r'C:\Users\songs\Desktop\Springboard Files\Springboard-Files\Take-home Challenges\ultimate_challenge\logins.json')
df.head()
```

```python
#Examining the data
df.info()
```

Okay, so we only have one column that's already in datetime format. First, let's resample the data by 15 minute intervals as the question had asked.

```python
df_15 = df.set_index('login_time')
df_15['count'] = 1
df_15 = df_15.resample('15min').sum()
df_15.head()
```

Now that we have this, we can look for some patterns in the login time. For example, we can look at the fluctuations that exist in **hours of the day.**

```python
#Making a copy of df_15
df_day = df_15

#Isolating each hour of the day and grouping by it
df_day['hour'] = df_15.index.hour
hourly_average = df_day.groupby('hour').mean()['count']
hourly_average
```

```python
#Plotting
plt.figure(figsize=(10,5))
plt.bar(range(24),hourly_average)
plt.xticks([0,3,6,9,12,15,18,21],['Midnight','3AM','6AM','9AM','Noon','3PM','6PM','9PM'],rotation=70)
plt.xlabel('Time of Day')
plt.ylabel('Mean Number of Logins')
plt.title('Daily Patterns in Login Time')
plt.show()
```

As is a bit unexpected of a transportation company, the peaks of login time do not correlate with rush hours (7-9AM and 4-6PM). In fact, those points see lower numbers of logins. Instead, the company sees its peak logins near lunch time and late night. The hour with the most average logins is 10PM, with an average of 16.2.

Next, let's look at the trends within each day of the week.

```python
#Making a copy of df_15
df_week = df_15

#Isolating each hour of the day and grouping by it
df_week['day'] = df_15.index.weekday
day_average = df_week.groupby('day').mean()['count']
day_average
```

```python
plt.figure(figsize=(10,5))
plt.bar(day_average.index,day_average)
plt.xticks(day_average.index,['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],rotation=70)
plt.xlabel('Day of Week')
plt.ylabel('Mean Number of Logins')
plt.title('Weekly Patterns in Login Time')
plt.show()
```

It appears that the number of logins start off low on Monday and gradually increases over the course of the week. As can be seen, weekends logins are greater than weekday logins.

Next we'll look at the monthly pattern.

```python
#Making a copy of df_15
df_month = df_15

#Isolating each hour of the day and grouping by it
df_month['day'] = df_15.index.day
daily_average = df_month.groupby('day').mean()['count']
daily_average
```

```python
#Plotting
plt.figure(figsize=(10,5))
plt.plot(daily_average)
plt.xticks([0,7,14,21,28])
plt.xlabel('Day of Month')
plt.ylabel('Mean Number of Logins')
plt.title('Monthly Pattern in Number of Logins')
plt.show()
```

A weekly pattern emerges: the number of logins start off high, then drops by the second or third day of the week, peaks again by the 4th day, drops again by the 5th or 6th day, and rises to another peak by the start of the next week. This is somewhat different from the weekly analysis we did earlier.

Generally, it appears that the mean number of logins stays constant over the passage of the month, with the higher peaks conteracted by lower drops.

Next, we'll look at the whole course.

```python
plt.figure(figsize=(10,5))
df_15['count'].rolling(960).mean().plot()
plt.xlabel('Date')
plt.ylabel('Rolling Mean (10 days) of Logins')
plt.title('Number of Logins over time')
plt.show()
```

4 times 15 is an hour, and 24 times an hour is a day. Thus, a period of 960 gives us a rolling mean of 10 days. I used a rolling mean because 1) It gives us more information about how the variable changes over time and specific patterns and breaks from the pattern compared to a bar plot, 2) There's a lot of points, and it'd be ineffective to plot them all.

Generally, there appears to be a seasonal pattern in the number of logins and an upward trend. There is a peak near the beginning of April, which could be looked at further. The actual peak might be earlier than suggested on the graph, because rolling mean has a tendency to push events back.

```python
url = r'C:\Users\songs\Desktop\Springboard Files\Springboard-Files\Take-home Challenges\ultimate_challenge\ultimate_data_challenge.json'
import json
with open(url) as file:
    data = json.load(file)
data = pd.DataFrame(data)
data.head()
```

#### The unique values in each columns
First, I take a look at all columns, the unique values in the categorical ones, a histogram of the numerical ones, and the existence of any null values.

```python
data.info()
```

I see that there are some null values in the "avg_rating_by_driver", "avg_rating_of_driver", and "phone" columns. I'll try to fill them in later, but for now I'll just look at the other columns, both to see if there's hidden null values and to get a better idea of the data. First, we look at the numerical columns.

```python
#We'll need a list of the numerical columns, they'd use useful for Linear Regression later
num_columns = []
for x in data.columns:
    if data[x].dtype == 'float64' or data[x].dtype == 'int64':
        plt.hist(data[x])
        plt.title(x)
        plt.show()
        num_columns.append(x)
```

It appears that most of the numerical columns, save for the weekday_pct, have a exponential distribution. Now, I'll look at the other columns. It'd be helpful to convert the dates to datetime objects, and look at the range of both the singup_date and the last_trip_date.

```python
#Converting columns to datetime
data['last_trip_date'] = data['last_trip_date'].astype('datetime64')
data['signup_date'] = data['signup_date'].astype('datetime64')
```

Now let's look at the range of the last_trip_date columns. It goes from January 1st to July 1st.

```python
print(data['last_trip_date'].min(),data['last_trip_date'].max())
```

Let's do the same for the signup_date column. It ranges from January 1st to January 31st, so all within a month.

```python
print(data['signup_date'].min(),data['signup_date'].max())
```

Now let's look at the remaining columns, phone and city.

#### Filling in NaN values

Next, we'd try to fill in some NaN values. First, I'm going to create a seaborn pairplot that, albeit messy, will give a idea of the relations between the variables that'd help.

```python
sns.pairplot(data[num_columns])
```

The first columns I will fill in are the numerical ones (avg_rating_by_driver, avg_rating_of_driver). I used Linear Regression to fill them in because, as seen in the pairplot above, both appear to have a linear relationship with the other numerical variables and a linear regression model will be best to capture these relationships.

```python
data[data.avg_rating_by_driver.isnull()].head()
```

```python
#Testing the model with not-null values

#Creating X(predictors) and y(target) variables
X = data[data['avg_rating_by_driver'].notnull()][num_columns].drop(['avg_rating_by_driver','avg_rating_of_driver'],axis=1)
y = data[data['avg_rating_by_driver'].notnull()]['avg_rating_by_driver']

#Train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

#Fit and predict linear regression
linear = LinearRegression()
linear.fit(X_train,y_train)
y_pred = linear.predict(X_test)
mean_absolute_error(y_test,y_pred)
```

```python
#Using the model to fill the missing values
X = data[data['avg_rating_by_driver'].isnull()][num_columns].drop(['avg_rating_by_driver','avg_rating_of_driver'],axis=1)
y_fill = linear.predict(X)

#Filling
indices = data[data['avg_rating_by_driver'].isnull()].index
data['avg_rating_by_driver'].loc[indices] = y_fill

#Test to see if there's any more.
data[data['avg_rating_by_driver'].isnull()]
```

```python
data[data.avg_rating_of_driver.isnull()].head()
```

```python
#Testing the model with not-null values

#Creating X(predictors) and y(target) variables
X = data[data['avg_rating_of_driver'].notnull()][num_columns].drop(['avg_rating_of_driver'],axis=1)
y = data[data['avg_rating_of_driver'].notnull()]['avg_rating_of_driver']

#Train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

#Fit and predict linear regression
linear = LinearRegression()
linear.fit(X_train,y_train)
y_pred = linear.predict(X_test)
mean_absolute_error(y_test,y_pred)
```

```python
#Using the model to fill the missing values
X = data[data['avg_rating_of_driver'].isnull()][num_columns].drop(['avg_rating_of_driver'],axis=1)
y_fill = linear.predict(X)

#Filling
indices = data[data['avg_rating_of_driver'].isnull()].index
data['avg_rating_of_driver'].loc[indices] = y_fill

#Test to see if there's any more.
data[data['avg_rating_of_driver'].isnull()]
```
