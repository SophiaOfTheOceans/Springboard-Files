# Classifying skin conditions based on pictures with machine learning

### Introduction

Every day, more than 9,500 people are diagnosed with skin cancer, making it the most common cancer in America. If left untreated, invasive cancerous cells can migrate to other parts of the body via the blood or lymph supply, and become lethal. As of 2011, the annual cost of treating skin cancer is $8.1 billion, and the Center for Disease Control (CDC) called it 'a growing public health issue'.

The most common type of skin cancer is melanoma, but other, lesser-know cancers such as intraepithelial carcinoma / Bowen's disease, basal cell carcinoma, and dermatofibroma also exist. All of them manifest as a darkened spot on the skin. However, many benign conditions of the skin present similarly, and for an untrained individual it may be difficult to differentiate between a such harmless condition and something more malign. 

In this capstone project, I would like to train a machine learning model using the HAM10000 dataset from the Medical University of Vienna (available at https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/DBW86T) and create an application of sorts for a non-health professional individual who suspects that he/she may have skin cancer. The individual can upload a picture of their skin condition, and the system will return the most likely diagnosis. Using this information, the individual can make a more informed decision on whether or not to pursue further treatment.

The dataset contains 10015 photos of cancer-like skin conditions and a .csv file with the lesion id, the image id, the diagnoses, the age of the patient, the sex of the patient, and the location of the condition.  The pictures are of Actinic keratoses and intraepithelial carcinoma / Bowen's disease (akiec), basal cell carcinoma (bcc), benign keratosis-like lesions (solar lentigines / seborrheic keratoses and lichen-planus like keratoses, bkl), dermatofibroma (df), melanoma (mel), melanocytic nevi (nv), or vascular lesions (angiomas, angiokeratomas, pyogenic granulomas and hemorrhage, vasc).

I plan to create a one-vs-all multi-class classification model to predict the diagnosis of a given picture as belonging to one of the 7 categories listed in Table 1. As the data is imbalanced with the majority diagnosis (6705 out of 10015) being “nv”, accuracy is not an effective metric; a possible alternative metric could be the confusion matrix or the precision-recall curve. In addition, an under-sampling technique to balance the distribution of the diagnoses might be necessary.

After performing feature extraction on the images and converting the information to tabular data, I intend to fit the data to kNN, Linear SVM, and Naive-Bayes, and compare the results.

For the deliverables, I will have code on GitHub, a report in PDF, and a jupyter notebook (this notebook right here) containing code.

First, we can look at the non-numerical columns, which seems to be the "lesion_id", "image_id", "dx", "dx_type", "sex", and "localization" columns.

```python
#What are the number of unique values in each non-numerical column?
for x in data.columns:
    if data[x].dtype == object:
        print(x + ": " + str(data[x].nunique()) + " unique values")
```

```python
#What exactly are the unique values in the categorical variables (objects with less than 20 unique values)?

#A list for simpler plotting of bar graphs later:
categoricals = []

#Print the values:
for x in data.columns:
    if data[x].dtype == object and data[x].nunique() < 20:
        categoricals.append(x)
        print("Unique values in the " + x + " column: " + str(data[x].unique()))
```
